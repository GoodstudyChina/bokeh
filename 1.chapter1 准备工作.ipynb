```python
from bokeh.plotting import figure, output_notebook, show
output_notebook()
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
p = figure(plot_width=400, plot_height=400)
p.circle(x, y, size=20, color="navy", alpha=0.5)
show(p)
```
